```python
# 7. Covert Columns From String ['sex' ,'smoker','region' ] To Numerical Values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
newGender = le.fit_transform(data['sex'])
newSmoker = le.fit_transform(data['smoker'])
newregion = le.fit_transform(data['region'])

data['sex'] = newGender
data['smoker'] = newSmoker
data['region'] = newregion
data
```

```python
# 8. Store Feature Matrix In X and Response(Target) In Vector y
```

```python

x = data[['age','sex','bmi','children','smoker','region']]
y = data['charges']
x
```

```python
# 9. Train/Test splitÂ¶
#1. Split data into two part : a training set and a testing set
#2. Train the model(s) on training set
#3. Test the Model(s) on Testing set# 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x,y,test_size = 0.33 , random_state = 5)
X_train
```

```python
# 10. Import the models
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# 11. Model Training
```

```python
model = LinearRegression()
model.fit(X_train,y_train)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train, y_train)

from sklearn.svm import SVR 
svm = SVR()
svm.fit(X_train, y_train)

from sklearn.ensemble import GradientBoostingRegressor
gr = GradientBoostingRegressor()
gr.fit(X_train, y_train)
```

```python
# 12. Prediction on Test Data
```

```python
y_pred = model.predict(X_test)
# df = pd.DataFrame({"Actual":y_test, "modelLR":y_pred})
# df
y_pred_rf = rf.predict(X_test)
# df = pd.DataFrame({"Actual":y_test, "rf":y_pred_rf})
# df
y_pred_svm = svm.predict(X_test)
# df = pd.DataFrame({"Actual":y_test, "svm":y_pred_svm})
# df
y_pred_gr = gr.predict(X_test)
df = pd.DataFrame({"Actual":y_test, "gr":y_pred_gr, "svm":y_pred_svm,  "rf":y_pred_rf, "modelLR":y_pred})
df
```
